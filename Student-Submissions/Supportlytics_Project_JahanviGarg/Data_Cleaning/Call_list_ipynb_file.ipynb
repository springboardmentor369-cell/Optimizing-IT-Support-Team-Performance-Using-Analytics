```python
import pandas as pd
```

```python
df = pd.read_excel("C:/Users/jahan/Downloads/Customer Call List.xlsx")
```

```python
#Dropping duplicates and useless column
df = df.drop_duplicates()
df = df.drop(columns="Not_Useful_Column")
```

```python
#Cleaning Last Name
df["Last_Name"] = df["Last_Name"].astype(str).str.strip("./_ ")
```

```python
#Cleaning Phone Number
df["Phone_Number"] = df["Phone_Number"].astype(str).str.replace(r'[^a-zA-Z0-9]', '', regex=True)
```

```python
def format_phone(number):
    if len(number) == 10:
       return f"{number[0:10]}"
    return ""

df["Phone_Number"] = df["Phone_Number"].apply(format_phone)
```

```python
#Standardize Yes/No
mapping = {'Yes': 'Yes', 'Y': 'Yes', 'No': 'No', 'N': 'No'}
df["Paying Customer"] = df["Paying Customer"].replace(mapping)
df["Do_Not_Contact"] = df["Do_Not_Contact"].replace(mapping)
df["Do_Not_Contact"] = df["Do_Not_Contact"].fillna("No")
```

```python
#Cleanup and Filtering
df = df.replace(["N/a", "NaN"], "")
df = df.fillna("")
df = df[df["Do_Not_Contact"] != "Yes"]
df = df[df["Phone_Number"] != ""]
```

```python
#Reset Index
df = df.reset_index(drop=True)
```
