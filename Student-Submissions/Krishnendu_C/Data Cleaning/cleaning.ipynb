```python
import pandas as pd
df=pd.read_csv(r'Technical Support Dataset.csv')
print(df)
```

```python
df=df.drop(columns=['Agent Name','Expected SLA to resolve','Expected SLA to first response','Close time','Survey results','Latitude','Longitude'])
df
```

```python
df=df.drop_duplicates()
df
print(df.dtypes)
```

```python
df['Created time'] = pd.to_datetime(df['Created time'])
df['First response time'] = pd.to_datetime(df['First response time'])
df['Resolution time'] = pd.to_datetime(df['Resolution time'])
df
```

```python
df['First response duration'] = (df['First response time'] - df['Created time']).dt.total_seconds() / 3600
df['Resolution duration'] = (df['Resolution time'] - df['Created time']).dt.total_seconds() / 3600
df
```

```python
df['SLA satisfied for first response'] = df['SLA For first response'].replace({'Within SLA': 'Yes', 'SLA Violated': 'No'})
df['SLA satisfied for resolution'] = df['SLA For Resolution'].replace({'Within SLA': 'Yes', 'SLA Violated': 'No'})
df.drop(columns=['SLA For first response','SLA For Resolution'], inplace=True)
df
```

```python
df= df.fillna(" ")
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
filename = 'Cleaned_Technical_Support_Dataset.csv'
df.to_csv(filename, index=False)
```
