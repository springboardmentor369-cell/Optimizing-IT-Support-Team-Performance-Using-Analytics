```python
import pandas as pd 
import numpy as np

df = pd.read_csv(
    r"C:\Users\balug\OneDrive\Documents\Desktop\Customer It Ticketing\customer_support_tickets.csv"
)
df
```

```python
df[["Customer Name", "Customer Email", "Ticket Priority", "Ticket Status"]].head()
df
```

```python
text_cols = ["Customer Name", "Customer Email", "Ticket Priority", "Ticket Status"]

for col in text_cols:
    df[col] = df[col].astype(str).str.strip()
df
```

```python
df[text_cols].head()
df
```

```python
df[["First Response Time", "Time to Resolution"]].head()
df
```

```python
df["First Response Time"] = pd.to_datetime(df["First Response Time"], errors="coerce")
df["Time to Resolution"] = pd.to_datetime(df["Time to Resolution"], errors="coerce")
df
```

```python
df[["First Response Time", "Time to Resolution"]].head()
df
```

```python
df["Resolution_Duration_Hours"] = (
    df["Time to Resolution"] - df["First Response Time"]
).dt.total_seconds() / 3600

df[["First Response Time", "Time to Resolution", "Resolution_Duration_Hours"]].head()
df
```

```python
df["Resolution_Duration_Hours"].describe()
df.loc[df["Resolution_Duration_Hours"] < 0, "Resolution_Duration_Hours"] = np.nan
df["Resolution_Duration_Hours"].describe()
df
```
